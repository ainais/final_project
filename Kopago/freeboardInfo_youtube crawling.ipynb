```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException

# 시간 명시
from tqdm import tqdm

# 오류 확인
import traceback
```

```python
keyword = '번역플랫폼'
url = 'https://www.youtube.com/results?search_query={}'.format(keyword)
```

```python
driver = webdriver.Chrome('./webdriver/chromedriver.exe')
driver.get(url)



time.sleep(5)
endk = 7
while endk:
    driver.find_element_by_tag_name('body').send_keys(Keys.END)
    time.sleep(0.3)
    endk -= 1


soup = bs(driver.page_source, 'html.parser')
#driver.close()
```

```python
name = soup.select('a#video-title')
video_url = soup.select('a#video-title')
view = soup.select('a#video-title')
videos = driver.find_elements_by_tag_name('ytd-video-renderer')
```

```python
# video 제목, url, 조회수, 썸네일
name_list = []
url_list = [] 
view_list = []
image_list = []

for i in video_url:
    url_list.append('{}{}'.format('https://www.youtube.com',i.get('href')))
    del url_list[33:92]
    
print(len(url_list))

for i in range(len(name)-(50)):
    name_list.append(name[i].text.strip())
    view_list.append(view[i].get('aria-label').split()[-1])

print(len(name_list))
print(len(view_list))
        
for video in videos:
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    image_list.append(thumbnail)
    image_list = list(filter(None,image_list))

print(len(image_list))

youtubeDic = {
    'videoName': name_list,
    'videoUrl': url_list,
    'videoView': view_list,
    'videoSrc':image_list
}
```

```python
youtubeDf = pd.DataFrame(youtubeDic)
youtubeDf
```
