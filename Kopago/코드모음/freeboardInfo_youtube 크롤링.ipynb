```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium.webdriver.common.keys import Keys
import time
```

```python
keyword = '번역플랫폼'
url = 'https://www.youtube.com/results?search_query={}'.format(keyword)
```

```python
driver = webdriver.Chrome('./webdriver/chromedriver.exe')
driver.get(url)



time.sleep(5)
endk = 5
while endk:
    driver.find_element_by_tag_name('body').send_keys(Keys.END)
    time.sleep(0.3)
    endk -= 1


soup = bs(driver.page_source, 'html.parser')
#driver.close()
```

```python
name = soup.select('a#video-title')
video_url = soup.select('a#video-title')
view = soup.select('a#video-title')

#videos = driver.find_elements_by_tag_name('ytd-video-renderer')
#len( videos )

 # 썸네일
#for video in videos[::43]:
 #   thumbnail = video.find_element_by_id('img').get_attribute('src')
  #  print(thumbnail)
```

```python
# video 제목, url, 조회수, 썸네일
name_list = []
url_list = []
view_list = []
image_list = []
```

```python
for i in range(len(name)):
    name_list.append(name[i].text.strip())
    view_list.append(view[i].get('aria-label').split()[-1])
    
for i in video_url:
    url_list.append('{}{}'.format('https://www.youtube.com',i.get('href')))
    
#for video in videos:
 #   thumbnail = video.find_element_by_id('img').get_attribute('src')
  #  image_list.append(thumbnail)
    
youtubeDic = {
    'videoName': name_list,
    'videoUrl': url_list,
    'videoView': view_list#,
   # 'videoSrc':image_list
}
```

```python
youtubeDf = pd.DataFrame(youtubeDic)
youtubeDf
```

```python
#youtubeDf.to_csv('C:/Users/haneul/youtube수정.csv', encoding='UTF-8', index=False)

#이 주피터 노트북 내 csv
youtubeDf.to_csv('youtubeInfo.csv', encoding='UTF-8', index=False)
```
