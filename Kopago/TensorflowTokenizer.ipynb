```python
import pandas as pd
df = pd.read_excel("conversation.xlsx")
df = df[df['대분류'] == '비즈니스']
df = df[df['소분류']=='회의']
df = df[df['상황'] == '의견 교환하기']
df
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

real_text = []
real_text.append(df['번역문'][0])
real_text.append(df['원문'][0])
real_result = [word_tokenize(sentence) for sentence in real_text]
print(real_result[0])
print(real_result[1])
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts(real_text)
print(tokenizer.word_index)

# 최종적으로 입력해야할 데이터 형식
print("===========================")
print("최종적으로 입력해야할 데이터 형식")
print(tokenizer.texts_to_sequences(real_text))
```
